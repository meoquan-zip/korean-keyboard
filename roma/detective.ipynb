## make labels

```python
from koreankb._load_data import *

I = load_initials()
M = load_medials()
F = load_finals()

def compose(_initial, _medial, _final=None) -> str:
    ini = I[_initial]
    med = M[_medial]
    fin = F[_final]
    return chr(0xAC00 + ini * 21 * 28 + med * 28 + fin)

ㅎ, ㅏ, ㅣ, ㄴ = 'ㅎ', 'ㅏ', 'ㅣ', 'ㄴ'
hanaz, haniz = [], []

for final in finrom:
    for initial in inirom:
        hana = compose(ㅎ, ㅏ, final) + compose(initial, ㅏ, ㄴ)
        hani = compose(ㅎ, ㅏ, final) + compose(initial, ㅣ, ㄴ)
        hanaz.append(hana)
        haniz.append(hani)
```

## filter

```python
hana_true = {}
hani_true = {}

with open('hana_true.txt', encoding='utf-8') as file:
    for line in file:
        hh, rr = line.strip().split(',')
        hana_true[hh] = rr

with open('hani_true.txt', encoding='utf-8') as file:
    for line in file:
        hh, rr = line.strip().split(',')
        hani_true[hh] = rr

len(hana_true), len(hani_true), 19 * 28
```

```python
hanaf = []
hanif = []
diff = []

for final in finrom:
    for initial in inirom:
        hana = compose(ㅎ, ㅏ, final) + compose(initial, ㅏ, ㄴ)
        hani = compose(ㅎ, ㅏ, final) + compose(initial, ㅣ, ㄴ)
        roma = f'ha{finrom[final]}{inirom[initial]}an'
        romi = f'ha{finrom[final]}{inirom[initial]}in'
        if roma != hana_true[hana]:
            hanaf.append(((final, initial), hana))
        if romi != hani_true[hani]:
            hanif.append(((final, initial), hani))
        if hana_true[hana][2:-2] != hani_true[hani][2:-2]:
            diff.append(((final, initial), (hana_true[hana][2:-2], hani_true[hani][2:-2])))

len(hanaf), len(hanif), len(diff)
```

```python
for xx in dict(hanif):
    if xx not in dict(hanaf):
        print(xx, dict(hanif)[xx])
```
